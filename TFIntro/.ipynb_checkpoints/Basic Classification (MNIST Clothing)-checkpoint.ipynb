# Basic Classification: Classify Images of Clothing

Source: [TF Tutorial](https://www.tensorflow.org/tutorials/keras/classification)

## Overview

Notes:
* The mnist fashion library contains 70k 28x28 greyscale images
* 60k train, 10k test images
* Outputs range from 0-9 and correspond to classes (t-shirt, trousers, dress, coat etc.)

```python
# tf and keras
import tensorflow as tf

# numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# load data
data = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = data.load_data()

# define class names
class_names = ['T-shirt/top', 'Trousers', 'Jumper', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Trainer', 'Bag', 'Boot']

# describe training features
desc_np_arr(x_train)
```

## Data Pre-Processing
Notes:
* We'll scale our pixel values down from greyscale (0-255) to 0-1 (ML models prefer these ranges)
* Then we'll check to see if our images correspond to the right labels

```python
# examine first training image
plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# scale pixel values (0-1 instead of 0-255)
x_train = x_train / 255.0
x_test = x_test / 255.0

# check scaled output
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

## Building the Model
Constructing:
* First layer flattens our 28x28 pixel images into a single, flat array of pixel values per image
* Dense layers are densely connected to the previous layer (i.e. each node in the previous layer connects to each node in the current layer)
* Dense layers also transform the dimensions of the input, so below where we specify "128", that is the dimension of the layer's output we will achieve (so we're essentially reducing our dimensions and extracting features)
* The final dense layer cuts our data's dimensions down to just 10, 1 for each output class
* The parameters of these layers (weights, biases etc.) will be adjusted during training but are random to begin with

Compiling:
* Our loss function determines how accurate our model is (actual vs. predicted outputs) and the aim is to minimize this loss function to improve our model
* The optimizer is the method used to adjust weights, biases, parameters etc. in order to reduce loss/error (e.g. adam, SGD, SGD with Nesterov momentum etc.)
* Metrics simply tell us how well our model has performed, there are different metrics to analyse models in different ways

```python
# construct our NN model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'), # 128 nodes/neurons
    tf.keras.layers.Dense(10)
])

# compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), # logits b/c we're not using softmax probabilities yet
              metrics=['accuracy'])
```

## Train the Model
Notes:
* Feed the training data into the model
* The model learns the relationships between inputs and outputs and adjusts it's parameters to optimize its ability to predict this
* Apply the model to the test data
* Validate the model to see how it performs (predicted labels vs. test/known labels)

```python
# fit model to training data
model.fit(x_train, y_train, epochs=10)
```

```python
# assess the performance of our model on the test data
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('\nTest Accuracy:', test_acc, '\nTest Loss:', test_loss)
```

## Making Predictions
Notes:
* It's worth noting before proceeding that our training accuracy was higher than our test accuracy
* This is often a sign of overfitting, where our model has essentially memorized our training data, making it a great predictor of the known images, but it struggles more with test data
* This means that it is not great at generalizing to new information despite being a strong predictor of the current, known information
* We could pause here and adjust our model to analyse, visualize and prevent overfitting but we'll do that in another notebook
* We will now make class predictions for specific input images using our trained model and visualize the output to see how well it's doing
* We'll add a softmax output layer to our model to enable us to analyse the meaningful class probabilities instead of log-odds

```python
# add softmax output layer to existing model
prob_model = tf.keras.Sequential([
    model,
    tf.keras.layers.Softmax()
])

# make predictions on test inputs
test_pred = prob_model.predict(x_test)

# examine first result
test_pred[0]
```

```python
# show us the index of the most probable class
np.argmax(test_pred[0])
```

```python
# it thinks this image is a boot
class_names[9]
```

## Analyse Results
Notes:
* Let's visualize our images against our predicted and known labels
* This will give us a clear indication of how effective our model is

```python
# plot image against actual and predicted label (with probability)
def plot_image(i, pred_arr, true_labels, images):
    # get image and actual label
    true_label, img = true_labels[i], images[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    # show image in greyscale
    plt.imshow(img, cmap=plt.cm.binary)
    
    # set blue if correct, red if incorrect
    pred_label = np.argmax(pred_arr)
    if pred_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    
    # plot image against actual and predicted label (with probability)
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[pred_label],
                                         100*np.max(pred_arr),
                                         class_names[true_label]),
                                         color=color)

# plot bar chart showing actual vs. predicted label with probability
def plot_value_array(i, pred_arr, true_labels):
    # get true label
    true_label = true_labels[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    
    # create bar plot
    thisplot = plt.bar(range(10), pred_arr, color='#777777')
    plt.ylim([0, 1])
    pred_label = np.argmax(pred_arr)
               
    # plot actual vs. pred (blue vs. red)
    # blue will overwrite/show if actual and pred are the same
    thisplot[pred_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
# show results for first test image
i = 0
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plot_image(i, test_pred[i], y_test, x_test)
plt.subplot(1, 2, 2)
plot_value_array(i, test_pred[i], y_test)
plt.show()
```

```python
# show results for second test image
i = 1
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plot_image(i, test_pred[i], y_test, x_test)
plt.subplot(1, 2, 2)
plot_value_array(i, test_pred[i], y_test)
plt.show()
```

```python
# plot x (specified) test images and highlight their actual vs. predicted labels
n_rows = 5
n_cols = 3
n_img = n_rows * n_cols
plt.figure(figsize=(2*2*n_cols, 2*n_rows))

# iterate through images
for i in range(n_img):
    plt.subplot(n_rows, 2*n_cols, 2*i+1) # 2 * n_cols b/c we have 2 plots per image, 2*i+1 calculate the subplot index
    plot_image(i, test_pred[i], y_test, x_test)
    plt.subplot(n_rows, 2*n_cols, 2*i+2)
    plot_value_array(i, test_pred[i], y_test)
plt.tight_layout()
plt.show()
```

## Making Single Predictions
Notes:
* Our model lets us extract a single image and make a prediction for it
* Keras models are optimized to make predictions on batches of data
* So even though we're just looking at one case here, we must add it to a batch/alter its dimensions

```python
# extract image, add to batch and predict class probabilities
img = x_test[1]
img = (np.expand_dims(img, 0)) # alter dimensions, create a batch with a single entry
single_pred = prob_model.predict(img)
print(single_pred)
```

```python
# visualize result
plot_value_array(1, single_pred[0], y_test)
_ = plt.xticks(range(10), class_names, rotation=45)
plt.show()
```

```python
# describe array
def desc_np_arr(arr):
    print("Shape =", arr.shape)
    print("Mean =", round(np.mean(arr), 1))
    print("Median =", round(np.median(arr), 1))
    print("Minimum =", round(np.amin(arr), 1))
    print("Maximum =", round(np.amax(arr), 1))
    print("Range =", np.ptp(arr))
    print("Variance =", round(np.var(arr), 1))
    print("Standard Deviation =", round(np.std(arr), 1))
```
